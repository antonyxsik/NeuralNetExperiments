Let's learn how to write a basic neural net

```julia
# packages were already installed by me in the Julia REPL
@time using Flux, Plots
```

```julia
# setup
data = [([x], 2x-x^3) for x in -2:0.1f0:2]

model = Chain(
    Dense(1 => 30, tanh), 
    Dense(30 => 1, 
    bias=false), only)

# setting optimizer to be Adam
optim = Flux.setup(Adam(), model)
```

```julia
#training 
for epoch in 1:500
    Flux.train!((m,x,y) -> (m(x) - y)^2, model, data, optim)
end
```

```julia
plot(x -> 2x-x^3, -2, 2, legend=false)
scatter!(x -> model([x]), -2:0.1f0:2)
```
